```python
import seaborn as sns
```

```python
df = sns.load_dataset("tips")
```

# HEATMAP


```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

# JOINTPLOT

```python
sns.jointplot(x = 'tip', y = 'size', data = df,kind = 'hex')
```

```python
sns.jointplot(x = 'tip', y = 'size', data = df,kind = 'reg')
```

# PAIR PLOT

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue = 'sex')
```

# DIST PLOT

```python
sns.distplot(df['total_bill'],kde = True,bins =5)
```

# CATEGORIAL PLOT

# COUNT PLOT

```python
sns.countplot(y = 'sex',data = df)
```

# BAR PLOT

```python
sns.barplot(y = 'total_bill',x = 'sex',data = df)
```

# BOX PLOT

```python
sns.boxplot(x = 'sex', y = 'total_bill', data = df, palette = 'rainbow')
```

```python
sns.boxplot(x = 'total_bill', y = 'day', data = df, hue = 'smoker')
```

# VIOLIN PLOT

```python
sns.violinplot(x = 'total_bill', y = 'day', data = df, hue = 'smoker')
```
