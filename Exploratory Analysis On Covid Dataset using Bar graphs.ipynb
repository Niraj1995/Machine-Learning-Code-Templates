```python
import pandas as pd
COVID_Data = pd.read_json("https://www.mohfw.gov.in/data/datanew.json")
COVID_Data
```

```python
COVID_Data = COVID_Data.iloc[:-1,1:]
```

```python
COVID_Data.head()
```

```python
import numpy as np
total_states = np.arange(len(COVID_Data['state_name']))
```

```python
total_states
```

```python
###### POSTIVE CASES BASED ON STATES
import matplotlib.pyplot as plt
plt.figure(figsize = (9,6),dpi = 60,facecolor='w',edgecolor='k')
plt.barh(total_states,COVID_Data['positive'],align='center',alpha=0.5,color = (0,1,0),edgecolor=(0.5,0.2,0.8))

plt.yticks(total_states,COVID_Data['state_name'])
plt.xlim(1,max(COVID_Data['positive'])+100)
plt.xlabel("No of Positive Cases")
plt.ylabel("Total States")
plt.title("Corona Virus Cases")
```

```python
###### NEW ACTIVE CASES BASED ON STATES
import matplotlib.pyplot as plt
plt.figure(figsize = (9,6),dpi = 60,facecolor='w',edgecolor='k')
plt.barh(total_states,COVID_Data['new_active'],align='center',alpha=0.5,color = (0,1,0),edgecolor=(0.5,0.2,0.8))
plt.yticks(total_states,COVID_Data['state_name'])
plt.xlim(1,max(COVID_Data['new_active']+10))
plt.xlabel("No of Positive Cases")
plt.ylabel("Total States")
plt.title("Corona Virus Cases")
```

```python
###### DEATH CASES BASED ON STATES
import matplotlib.pyplot as plt
plt.figure(figsize = (9,6),dpi = 60,facecolor='w',edgecolor='k')
plt.barh(total_states,COVID_Data['death'],align='center',alpha=0.5,color = (0,1,0),edgecolor=(0.5,0.2,0.8))
plt.yticks(total_states,COVID_Data['state_name'])
plt.xlim(1,max(COVID_Data['death']+10))
plt.xlabel("No of Positive Cases")
plt.ylabel("Total States")
plt.title("Corona Virus Cases")
```
