# For String Data CSV

```python
from io import StringIO, BytesIO
```

```python
data = ('col1,col2,col3\n'
'x,y,1\n'
'a,b,2\n'
'c,d,3')
```

```python
data = pd.read_csv(StringIO(data),usecols=['col3'])
```

```python
data.head()
```

```python
data = ('col1,col2,col3\n'
'1,2,3\n'
'4,5,6\n'
'7,8,9')
```

```python
print(data)
```

```python
data = pd.read_csv(StringIO(data),dtype = object)
```

```python
data['col1'][1]
```

```python
data = pd.read_csv(StringIO(data),dtype = int)
```

```python
print(data['col1'][1])
```

```python
import numpy as np
```

```python
data = pd.read_csv(StringIO(data),dtype = {'col1':int,'col2':np.float,'col3':np.float})
```

```python
data = pd.read_csv(StringIO(data),index_col = 0)
data
```

```python
data = pd.read_csv(StringIO(data),index_col = False)
data
```

```python
data = 'a,b\n"hello, \\"Bob\\",nice to see you",5'
pd.read_csv(StringIO(data),escapechar = '\\')
```
