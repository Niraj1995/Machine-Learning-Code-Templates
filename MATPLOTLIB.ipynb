```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
x=np.arange(40,50)
y=np.arange(50,60)
```

# Scatter plot

```python
plt.scatter(x,y,c = 'g')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("2-D graph")
plt.savefig("Test.png")
```

```python
y = x*x
```

```python
plt.plot(x,y,'r')
```

```python
plt.plot(x,y,'r--')
```

```python
plt.plot(x,y,'r*-')
```

```python
plt.plot(x,y,'ro',linestyle = 'dashed',linewidth = 2,markersize = 12)
```

```python
### Creating Subplot

plt.subplot(2,2,1)
plt.plot(x,y,'r')
plt.subplot(2,2,2)
plt.plot(x,y,'r')
```

```python
x = np.arange(20,30)
y = 3 * x * 5
plt.plot(x,y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Matplotlib diagram")
```

```python
x = np.arange(0,4*np.pi,0.2)
y = np.sin(x)
plt.plot(x,y)
```

```python
x = np.arange(0,4*np.pi,0.2)
y_sin = np.sin(x)

plt.subplot(2,1,1)
plt.plot(x,y_sin)

x = np.arange(0,4*np.pi,0.2)
y_cos = np.cos(x)

plt.subplot(2,1,2)
plt.plot(x,y_cos)
```

# BAR PLOT

```python

x = [2,8,10]
y = [11,16,19]

x2 = [3,9,11]
y2 = [6,15,7]

plt.bar(x,y,color = 'r')
plt.bar(x2,y2,color = 'g')

plt.show()
```

# BOX PLOT

```python
data = [np.random.normal(0,std,100) for std in range(1,4)]

plt.boxplot(data,vert = True,patch_artist = True)
```

# PIE CHART

```python
sizes = [215,130,541,350]
label = "Python","Ruby","C++","R"
colors = ['yellow','green','pink','red']
explode = (0.1,0,0,0)

plt.pie(sizes,labels = label,colors = colors,explode = explode,autopct ='%1.1f%%',shadow=True)

plt.axis('equal')
plt.show()
```
